# CAN NOW BEGIN TO CALCULATE VARIOUS BACKTESTING STRATEGIES SUCH AS RELATIVE STRENGTH INDEX, SIMPLE MOVING AVERAGES, AND EXPONENTIAL MOVING AVERAGES 


# SIMPLE MOVING AVERAGE IS CALUCLATED BY 

SMA of time N = ( P1 + P2 + P3 + ... + Pn ) / n 

where P1,P2,P3,Pn are prices over the last n periods  

N is the window size (ex: 14 days, 50 days)

_______________________________________________

# EMA (Exponential Moving Average)

EMA of time t = α * Pt + ( 1 - α ) * EMAt-1 
where Pt is the price at time t, EMAt-1 is the EMA from previous time

 α = 2/ n + 1 which is the smoothing factor
 
n = number of periods (ex: 14 days, 50 days)

NOTE: EMA gives more 'weight' to current/recent prices

_______________________________________________

# RIS relative strength index
Price change = delta P 
Gain = average of positive changes over n periods
Loss = average of negative changes over n periods

RS = average Gain / average Loss

RSI = 100 - (100 / 1 + RS) which always ranges from 0 to 100

RSI > 70 --> overbought

RSI < 30 --> oversold 

```python
# calculates the SMA over 14 days 
#
hist_OHLC_cleaned['SMA_14'] = hist_OHLC_cleaned['Close_NVDA'].rolling(window = 14).mean()

```

```python
# calculates EMA over 14 days 
# ewm is the exponential weighted moving average function listed above
hist_OHLC_cleaned['EMA_14'] = hist_OHLC_cleaned['Close_NVDA'].ewm(span = 14, adjust = False).mean()
```

```python
# in order to compute RSI we need to create a helper function 
def compute_rsi(series, period = 14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window = period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window = period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

hist_OHLC_cleaned['RSI_14'] = compute_rsi(hist_OHLC_cleaned['Close_NVDA'])

#diff calculates the difference between the current and previous value 
#delta.where delta > only keeps the values where the condition is true (positive changes in price) otherwise it replaces with 0
#delta.where delta < only keeps negative changes in price otherwise replace with 0
#then... we can calculate the rs and rsi values to return

```

WE WILL NOW BE IMPLEMENTING A MULTI LINE SMA STRATEGY TO SEE HOW WE CAN DEVELOP BUY AND SELL SIGNALS WHICH WILL INCLUDE:

- SHORT TERM SMA AND LONG TERM SMA

- BUY SIGNAL : SHORT SMA CROSSES ABOVE LONG TERM SMA

- SELL SIGNAL : SHORT SMA CROSS BELOW THE LONG TERM SMA

- FURTHER IMPLEMENTATION: TAKE PROFIT AND STOP LOSS EXIT TARGETS BASED OFF PERCENTAGE 

```python
import pandas as pd
import yfinance as yf

end_date = datetime.today()
print("End Date:", end_date)

start_date = end_date - timedelta(days = 2 * 365)  # Example: 2 years ago
print("Start Date:", start_date)


df = yf.download("NVDA", start = start_date, end = end_date)
df = df[['Close']] # we will only use the closing price of a given ticker

#compute the long and short term smas
df['SMA_20'] = df['Close'].rolling(window = 20).mean()
df['SMA_50'] = df['Close'].rolling(window = 50).mean()

```

```python
# Now we can generate the buy and sell signals for the stocks

#STEP 1: CREATE NEW DATAFRAME HOLDING THE SIGNAL VALUE
df['Signal'] = 0 #1 will mean buy/hold position, 0 will mean dont buy/not holding

#STEP 2: COMPARE SMA VALUES 
#this mean
#Set the values in the 'Signal' column of the dataframe, starting at row 50(day 50) and onward, to 1 if SMA_20 is greater than SMA_50, or 0 if not.
# we use indexLoc (iloc) at 50 onward because the long term moving average window is 50th day exclusive and above have real numbers
# if we were to use iloce at 20 onward only the short term sma would have valid numbers
df['Signal'] = 0
df.iloc[50:, df.columns.get_loc('Signal')] = (df['SMA_20'] > df['SMA_50']).astype(int).iloc[50:]




#STEP 3: CREATE NEW POSITION COLUMN HOLDING WHAT WE ARE GOING TO DO BASED OFF SIGNAL +1:BUY , -1:SELL, 0:DO NOTHING
#.diff() would calculate the previous value with current value creating and "offset" which would directly corelate to our buy and sell signals
df['Position'] = df['Signal'].diff()

```

```python
#we can now plot the different moving averages, close price, and buy/sell signals
import matplotlib.pyplot as plt

#create figure
plt.figure(figsize=(14,6))

#plot close prices
plt.plot(df['Close'], label = 'Close Price', color = 'black', linewidth = 1)

#plot the SMA lines
plt.plot(df['SMA_20'], label = 'SMA 20', color = 'blue', linestyle = '--')
plt.plot(df['SMA_50'], label = 'SMA 50', color = 'orange', linestyle = '--')

#plot buy signal (position == 1)
#this means
#filtered df by position == 1 (buy signal) and its respective date (.index) --> plot the df col values only where the df col 'Position' equals 1
#get close price at each position == 1
#mark that index with a ^ indicating buy
#markersize 10 and color for visibility
plt.plot(df[df['Position'] == 1].index,
         df['Close'][df['Position'] == 1],
         '^', markersize = 10, color = 'green', label = 'Buy Signal')

#plot sell signal (position == -1)
plt.plot(df[df['Position'] == -1].index,
         df['Close'][df['Position'] == -1],
         'v', markersize = 10, color = 'red', label = 'Sell Signal')

#add final chart settings
plt.title('SMA Crossover Strategy on NVIDIA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()              #helps delete the white space between different labels 
plt.show()
```

 WE WILL NOW BEGIN TO SIMULATE THE STRATGY'S RETURN VS SIMPLY HOLDING NVDA

 WHAT WE HAVE TO DO:
 
 - CALCULATE DAILY RETURNS FOR NVDA (BUY AND HOLD BASELINE TO COMPARE TO)

 - APPLY SIGNALS WE CREATED TO GET STRATEGY RETURNS

 - TRACK TOTAL RETURN OVER TIME

 - PLOT THE CURVES COMPARING EACH OTHER

 - REMEMBER BASIC STRATEGY ASSUMPTIONS THAT SIGNAL == 1 YOU OWN THE STOCK, SIGNAL == 0 NO RETURNS THAT DAY

 - RETURNS ARE CALCULATED USING PERCENT CHANGES SO WE CAN GO BASED OFF ANY ACCOUNT SIZE 

```python
# CALCULATE DAILY % RETURNS 
df['Market Return'] = df['Close'].pct_change() 

# only apply return to strategy when the signal was triggered and you were "in" the trade 
df['Strategy Return'] = df['Market Return'] * df['Signal'].shift(1) #shift 1 day (down 1 row)
```

```python
#CALCULATE CUMMULATIVE GROWTH
principle = 1

df['Cumulative Market Return'] = (principle + df['Market Return']).cumprod()
df['Cumulative Strategy Return'] = (principle + df['Strategy Return']).cumprod()

# since returns are percent based, multiplying them together obtains the actual value therefore we use cumprod (cumulative product)
# we have to have an initial investment (our principle) so we add that as well 
# we wouldnt multiply the principle because our returns are percent based so we have to scale it up as our principle is being updated 
# each step would compound off the previous day 
```

```python
# NOW WE CAN PLOT THE RESULTS
plt.figure(figsize = (14,6))
#plt.plot(df['Cumulative Market Return'],label = 'Buy and Hold NVDA')
#plt.plot(df['Cumulative Strategy Return'],label = 'SMA Crossover Strategy')

# could also do percent return
plt.plot((df['Cumulative Strategy Return'] - 1) *  100, label = 'SMA Crossover Strategy')
plt.plot((df['Cumulative Market Return'] - 1) *  100, label = 'Buy and Hold NVDA')

plt.title('Returns of SMA Strategy vs Buy and Hold NVDA')
plt.xlabel('Date')
plt.ylabel('Growth with principle (% RETURN)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

NOW WE CAN BEGIN IMPLEMENTING THE TAKE PROFIT AND STOP LOSS LOGIC 

WE NEED TO ACCOUNT FOR USER INPUT ULTIMATELY 
- TAKE PROFIT ONCE REACHED A CERTAIN TARGET %
- STOP LOSS ONCE LOSS EXCEEDS A SPECIFIC %

LOGIC NEEDED
- IF PRICE INCREASE ENOUGH TRIGGER TAKE PROFIT AND SELL
- IF PRICE DROPS TOO MUCH TRIGGER STOP LOSS AND SELL 
- OTHERWISE HOLD UNTIL SMA TELLS YOU TO EXIT

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

#create start and end dates
numYears = 2
end_date = datetime.today()
print("End Date:", end_date)

start_date = end_date - timedelta(days = numYears * 365)  # Example: 2 years ago
print("Start Date:", start_date)


# FIND OUR PARAMETERS FOR TAKE PROFIT AND STOP LOSS (IN PERCENT)

take_profit_pct = 0.30 #20% gain
stop_loss_pct = 0.05 #10% loss

#flatten the columns because they are multi indexed
# This will flatten the MultiIndex columns
df_final.columns = df_final.columns.get_level_values(0)
```

```python
#loop through our data because now we need to dynamically react to how price moves 
# need to check if we are in a trade, if we need to exit, remember entry price, calculate gain with exit price

tp_sl_signals = [] #store each signal for each day
in_trade = False
entry_price = 0

for i in range(len(df_final)):
    close_price = df_final['Close'].iloc[i]#current day's closing price

    #print("Row:", i)
    #print("Signal:", df_final['Signal'].iloc[i])
    #print("In Trade?", in_trade)


    if not in_trade:
        #entry condition with SMA signal
        if df_final['Signal'].iloc[i] == 1: 
            entry_price = close_price #since we only look at close prices, our entry price would be saved as the current close price
            in_trade = True
            tp_sl_signals.append(1) #append 1 to our list to indicate that we are currently in a trade
        else:
            tp_sl_signals.append(0) #append 0 to indicate we arent in a trade
    else:
        #exit conditions where SMA crosses over or lose enough percent
        current_return = (close_price - entry_price) / entry_price
        sma_exit = df_final['Signal'].iloc[i] == 0 #SMA crossed so tells us to exit

        #print("Return:", current_return)
        #print("SMA Exit:", sma_exit)
        #print("TP:", current_return >= take_profit_pct)
        #print("SL:", current_return <= -stop_loss_pct)


        #exit trade if we hit tp, or if we hit sl or if price never hits tp/sl and SMA crosses over we have to exit
        if current_return >= take_profit_pct or current_return <= -stop_loss_pct or sma_exit:
            in_trade = False
            tp_sl_signals.append(0)
        else:
            tp_sl_signals.append(1)
```

```python
#add signals to the dataframe

df_final['TP_SL_Signal'] = tp_sl_signals
```

```python
# Compute strategy returns

#daily return of the stock
df_final['Market Return'] = df_final['Close'].pct_change()

#strategy return only when the signal is on
df_final['Strategy Return'] = df_final['Market Return'] * df_final['TP_SL_Signal'].shift(1)

#calculate the growth over time
initial_capital = 1

df_final['Cumulative Market Return'] = (1 + df_final['Market Return']).cumprod() * initial_capital
df_final['Cumulative Strategy Return'] = (1 + df_final['Strategy Return']).cumprod() * initial_capital
```

```python
# can now plot the curves to compare
import matplotlib.pyplot as plt

#changing values to percents
#df_final['Cumulative Market Return %'] = (df_final['Cumulative Market Return'] - 1) * 100
#df_final['Cumulative Strategy Return %'] = (df_final['Cumulative Strategy Return'] - 1) * 100


plt.figure(figsize=(14,6))
plt.plot(df_final['Cumulative Market Return'], label = 'Buy and Hold', linestyle = '--', color = 'orange')
plt.plot(df_final['Cumulative Strategy Return'], label = 'TP/SL/SMA Signals', color = 'blue')
plt.title('Strategy vs. Market Performance')
plt.xlabel('Date')
plt.ylabel('Growth of 1$ Investment (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
